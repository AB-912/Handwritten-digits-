```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
dataset = load_digits()
```

```python
print(dataset.data)
print(dataset.target)

print(dataset.data.shape)
print(dataset.images.shape)

dataimlen = len(dataset.images)
print(dataimlen)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1500

plt.gray()
plt.matshow(dataset.images[n])
plt.show

dataset.images[n]
```

```python
X = dataset.images.reshape((dataimlen,-1))
Y = dataset.target
print(X)
print()
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state=0)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn import svm
```

```python
model = svm.SVC(gamma = 0.001, C =0.45)
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
cm = confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
print("THE ACCURACY OF THE MODEL FOR DETECTING HANDWRITTEN DIGITS : {0}%".format(accuracy_score(Y_test,Y_pred)*100))
```

```python
n = 888
result = model.predict(dataset.images[n].reshape((1,-1)))
plt.imshow(dataset.images[n], cmap=plt.cm.gray_r, interpolation='nearest')
print(result)
print("\n")
plt.axis('off')
plt.title('%i' %result)
plt.show()
```
